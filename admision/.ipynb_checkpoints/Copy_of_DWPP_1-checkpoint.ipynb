```python
# First I'll import the modules to be used in this assignment 
from sklearn.datasets import load_iris
from typing import Iterable,Tuple,Sequence,Dict,List
from collections import defaultdict
from math import fsum,sqrt
from pprint import pprint
import random
from functools import partial
import matplotlib.pyplot as plt
#The seed is fixed for reproducibility
random.seed(4200)
# The next thing to do is to get the dataset, to do that we just used the sklearn library which allows one to import the dataset in just one command
data=load_iris()
```

```python
# Next I'll define some types for my data annotation
Point=Tuple[float,...]
Centroid=Point
```

```python
# This just transforms the data of the data set into Points
data_points=list(map(tuple,data['data'].tolist()))
```

# Next I'll implement the K-means clustering algorithm, to do that I'll implement the following steps:
- Sample K random points from my data and use as my initial centroids
- Calculate the distance of each point to each centroid
- Assign each point to the group of it's closest centroid
- Compute the centroids (the mean of each group)



```python
def mean(data:Iterable[float])->float:
    'Accurate Aritmethic mean'
    data=list(data)
    return fsum(data)/len(data)
```

```python
22.2-1.1
```

```python
def distance(p:Point,q:Point,fsum=fsum,sqrt=sqrt,zip=zip)->float:
    'Euclidean distance in N dimensional Space'
    return sqrt(fsum([(x-y)**2 for x,y in zip(p,q)]))
```

```python
def assign_groups(centroids:Sequence[Centroid],data:Iterable[Point])->Dict[Centroid,List[Point]]:
    'Group the data points to the closest centroid'
    d=defaultdict(list)
    for point in data:
        closest_centroid=min(centroids,key=partial(distance,point))
        d[closest_centroid].append(point)
    return dict(d)
```

```python
def compute_centroid(groups:Iterable[Sequence[Point]])->List[Centroid]:
    'Compute centroid of each group'
    return [tuple(map(mean,zip(*group))) for group in groups]
```

```python
def k_means(data,k=2,iterations=300):
    'Implements the K-means algorithm'
    centroids=random.sample(data,k)
    for i in range(iterations):
        labeld=assign_groups(centroids,data)
        centroids=compute_centroid(labeld.values())
    return centroids,labeld
```

```python
centers,labeled=k_means(data_points,k=3)
```

# Now let us check our implementation of K-means by comparing the output with the implementation from the popular library sklearn

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(data['data'])
```

```python
x=data['data']
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'b', label = 'Cluster 1')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'r', label = 'Cluster 2')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```

```python
centers_sk=kmeans.cluster_centers_.tolist()
```
